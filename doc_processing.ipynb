```python
from google import genai 
import langextract as lx 
from dotenv import load_dotenv
import os
```

```python
def llm_respond(Content):
    load_dotenv()
    key = os.getenv("API_KEY")
    

    client = genai.Client(api_key=key)
    response = client.models.generate_content(
        model = "gemini-3-flash-preview" , contents = Content )
    return print(response.text)

llm_respond("explain in detals the how diffusion models works in 3 lignes")
```

```python
examples = [
    lx.data.ExampleData(
        text="Adel is the Hr manager at deloitte ",
        extractions = [
            lx.data.Extraction(
                extraction_class = "hr_name",
                extraction_text = "adel"
            )

        ]
    )
]
```

```python
class PDFProcessing :
    def __init__(self,text,example):
        self.input = text
        self.examples = example
        self.API_KEY = None


    

    def get_apikey(self):

        load_dotenv()
        key = os.getenv("API_KEY")
        self.API_KEY = key
        return  self.API_KEY

    def processing_pdf(self):

        API_KEY = self.get_apikey()
        result = lx.extract(
            text_or_documents=self.input,
            prompt_description="extract the name of the Hr from this text  ",
            examples=examples,
            model_id = "gemini-3-flash-preview" ,
            api_key = API_KEY
        )
        return result
        
```

```python
text = "adel is the HR of the deloitte "

pp = PDFProcessing(text=text,examples)
result = pp.processing_pdf()
print(result)
```
